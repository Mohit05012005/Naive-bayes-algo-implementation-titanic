```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/titanic.csv')
```

```python
data
```

```python
data.head()
```

```python
# name and those column's which  do not have any impact on the output we will drop it out.
data.drop(['Name','Sex','Age','SibSp','Parch','Ticket','Cabin','Embarked'],axis=1,inplace=True)
```

```python
data.head()
```

```python
data.info()
```

```python
data.dropna()
```

```python
data['Fare'] = data['Fare'].fillna(data['Fare'].mean())
```

```python
data.columns[data.isna().any()]  # to check wheather there is any null value in any of the columns if yes then it will give the column
```

```python
input = data[['Fare','Pclass','PassengerId']]
output = data['Survived']
```

```python
input
```

```python
output
```

```python
# now divide the data into test data and train data
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(input,output,test_size=0.2,random_state=24)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
model = gnb.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
predict = model.predict(x_test)
```
